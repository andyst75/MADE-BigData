```python
from collections import Counter

import pandas as pd

df = pd.read_csv('artists.csv', low_memory=False)

tags = []
no_tags = 0

for item in df.tags_lastfm.to_list():
    if isinstance(item, float):
        no_tags += 1
    else:
        for x in item.split(';'):
            tags.append(x.strip())
        
print('No tags:', no_tags)
Counter(tags).most_common(10)
```
